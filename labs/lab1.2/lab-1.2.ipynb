## Pyplot  
 
* matplotlib.pyplot is a collection of functions that make matplotlib work like MATLAB. 
* Each pyplot function makes some change to a figure: e.g., 
  * creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.
* In matplotlib.pyplot states are preserved across function calls.
  * This means it keeps track of things like the current figure and plotting area
  * The plotting functions are directed to the current axes 



## Figures and axes <a id='H8'></a>

In matplotlib there are two important object types, the figure object and the axes object. 

* This sometimes leads to confusion, because you will see code that uses axes objects commands and sometimes not.

* So what is the difference between these two objects?

* **Figures** (think of this as your "canvas")
  * Think of figures as the blank canvas (or paper) on which you are going to put your plots (i.e. the whole figure).
  * The Figure keeps track of all the child Axes, a group of attributes (titles, figure legends, colorbars, etc), and even nested subfigures.
  * It is often convenient to create the Axes together with the Figure, but you can also manually add Axes later on. 
  * When you modify the figure attribute you are modifying the canvas rather than the subplots


* **Axes** (individual plots)
  * Think of axes as the individual plots that go on the paper.
    * Each figure (canvas) can contain one or more Axes. 
      * Just like with pen and paper, you can put one plot on the paper or many (i.e. subplots)
    * The axes are areas where points can be specified in terms of x-y coordinate system 
      * (or theta-r in a polar plot, x-y-z in a 3D plot, etc). 
    * The simplest way of creating a Figure with an Axes is using pyplot.subplots. 
    * We can then use Axes.plot to draw some data on the Axes:
    * The Axes class and its member functions are the primary entry point to working with the MLP interface
      * They have most of the plotting methods defined on them (e.g. ax.plot())

The following figure exemplifies the difference between the two types of objects.

![](https://files.realpython.com/media/fig_map.bc8c7cabd823.png)

Notice that we can have multiple axes objects per figure. 

![](https://files.realpython.com/media/empty_gridspec_annot.782349dc90a1.png)

**Because of these two object types, there are essentially two ways to use Matplotlib:**
  * Method-1) Rely on pyplot to automatically create and manage the Figures and Axes, and use pyplot functions for plotting.
  * Method-2) Explicitly create Figures and Axes, and call methods on them (i.e. the "object-oriented (OO) style").

### Method-1  

* In this case we don't explicitly define the figure and axes object
* However, the code implicitly defines them in the background


````python
import matplotlib.pyplot as plt
import numpy as np

#DEFINE DATA
x=[1, 2, 3, 4]
y=[1, 2, 3, 4]

#PLOT
plt.plot(x,y,'o') # 'o' makes a scatterplot or point plot
plt.show()
````

**Basic customization**

* We can still do some basic customization, even though we did not explicitly define the figure and axes objects. 
* These commands are modifying attributes of the implicitly defined figure object.

````python
import matplotlib.pyplot as plt

#DEFINE DATA
x=[1, 2, 3, 4]
y=[1, 2, 3, 4]

#PLOT
plt.plot(x,y,'-o') # '-o' makes plots points and a line

#SET FIGURE LABELS AND TITLE 
FS=12	 
plt.suptitle("Newton's Law: F=ma", fontsize=FS)
plt.xlabel('Acceleration (m/s/s)', fontsize=FS)
plt.ylabel('Force (N)', fontsize=FS)
#TUPLE SPECIFIES WHERE TO PUT ANNOTATION
plt.annotate("slope=mass (kg)", (1,3.65), fontsize=FS) 
plt.show()
````

### Method-2 

* In this case we explicitly define both a figure object and an axes object.
* This method is preferred, because it is more explicit, and it generalizes to the case of subplots, as we will see later.

````python
import matplotlib.pyplot as plt
import numpy as np

#DEFINE DATA
x=[1, 2, 3, 4]
y=[1, 2, 3, 4]

#DEFINE OBJECTS 
fig, ax = plt.subplots()

#PLOT
ax.plot(x,y,'o')
plt.show()
````

**Basic customization** 
* Just as with method-1, we can do various forms of customization.
* However notice that some of the commands are different, since we are acting on the axis object.

````python
import matplotlib.pyplot as plt
import numpy as np

#DEFINE DATA
x=[1, 2, 3, 4]
y=[1, 2, 3, 4]

#DEFINE OBJECTS 
fig, ax = plt.subplots()

#PLOT
ax.plot(x,y,'-o')

#CUSTOMIZE
FS=12	 
ax.set_title("Newton's Law: F=ma", fontsize=FS)
ax.set_xlabel('Acceleration (m/s/s)', fontsize=FS)
ax.set_ylabel('Force (N)', fontsize=FS)
ax.annotate("slope=mass (kg)", (1,3.65), fontsize=FS) 
plt.show()
````

## Customizing FacetGrid objects

However, you can typically access the underline MLP objects via the attributes of the returned object, as shown in the following example.

This allows you to customize a Seaborn plot just like you would a regular MPL plot. 

````python
g=sns.relplot(
data=df, 
x="total_bill", 
y="tip", 
hue="sex",
col="time",
s=200,
)

#NOTICE THAT <class 'seaborn.axisgrid.FacetGrid'> OBJECT HAS ATTRIBUTES CONTAINING MPL OBJECTS
print('-----TYPES-----')
print(type(g.fig))
print(type(g.axes[0,0]))
print(type(g.axes[0,1]))

# BY ACCESSING THE MPL AXES/FIG ATTRIBUTES, WE CAN USE ALL OF OUR FAMILIAR MPL COMMANDS TO CUSTOMIZE

FS=16

#FIGURE TITLE
g.fig.set_size_inches(12, 6)

#AXIS LABELS AND FONT SIZE
g.axes[0,0].set_xlabel('Total bill at restaurant per table (US dollars)', fontsize=FS)
g.axes[0,0].set_ylabel('Total tip per table (US dollars)', fontsize=FS)
g.axes[0,1].set_xlabel('Total bill at restaurant per table (US dollars)', fontsize=FS)

#TITLE FONT SIZE
g.axes[0,0].set_title("Lunch time",fontsize=FS)
g.axes[0,1].set_title("Dinner time",fontsize=FS)


#AXIS TIC FONT SIZE
for i in range(0,2):
    g.axes[0,i].tick_params(axis='both', which='major', labelsize=FS)
    #g.axes[0,i].tick_params(axis='both', which='minor', labelsize=FS)

#LEGEND FONT SIZE
g.axes[0,0].legend(fontsize=FS)

#REMOVE SIDE LEGEND
g._legend.remove()

plt.show()
````

# Theme creation

Typically the default properties of plots are un-acceptable for professional level presentations. 

This is because, the labels are not very intuitive, lack units, and have font that is too small.

Therefore, we often want to modify them to generate higher-quality visualizations. 

**The default parameters of MPL are stored in an object called rcParams which is similar to a dictionary**

**Therefore, there are essentially four ways to customize Matplotlib:**

* 1) Modifying rcParams at runtime.
* 2) Using style sheets.
* 3) Changing your matplotlibrc file.
* 4) Manually modifying attributes of your plot when you create it (as shown in the simple example above)

**NOTE**: Setting rcParams at runtime takes precedence over style sheets, style sheets take precedence over matplotlibrc files.

**NOTE:**: If you change the parameters in rcParams they will be re-set when you restart the python kernel.

You can print the defaults in the rcParams object, as shown below. 

## rcParams file 

`matplotlib` has a styling system that is a bit more robust, using a construct called `rcParams`

When you use a style in `matplotlib`, what happens under the hood is that various elements of `matplotlib.rcParams` are changed to meet the specifications.

You can manipulate `matplotlib.rcParams` yourself as well.

````python
# THIS IS IMPORTANT: THERE ARE 308 ENTRIES IN rcParam 
# WHICH SHOWS YOU MOST OF WHAT YOU CAN CUSTOMIZE WITH MPL
print(plt.rcParams)
````

### Plot with default rcParam

lets start with a default mpl plot 

````python
def mpl_plot():
    # DATA FOR PLOTTING
    t = np.arange(0.0, 2.0, 0.01)
    s = 1 + np.sin(2 * np.pi * t)

    # INITIALIZE 
    fig, ax = plt.subplots()

    # PLOT
    ax.plot(t, s)
    plt.show()

mpl_plot()
````

````python
# LETS START WITH A GENERIC DEFAULT SEABORN PLOT

def sns_plot():
    tips = sns.load_dataset("tips")  # USE DEFAULT DATA-SET INCLUDED WITH PYTHON 
    sns.histplot(tips['total_bill'])
    plt.show()

sns_plot()
````

### Editing rcParam

This modification will effect all future plots, unless we change it back. 

Or until we restart the python kernel, e.g. by reopening the .ipynb file or re-running the .py file

This will restore the un-modified rcParam defaults. 

````python
# print(type(plt.rcParams))
print("BEFORE:", plt.rcParams["figure.figsize"])

# YOU CAN ALSO MODIFY THESE ATTRIBUTES
plt.rcParams["figure.figsize"]= (10,10)     # change the default figure size to a 10in x 10in resolution
print("AFTER:",plt.rcParams["figure.figsize"])

plt.rcParams["axes.grid"]=True
plt.rcParams["axes.linewidth"]=2
plt.rcParams["lines.linewidth"]= 4
plt.rcParams["lines.linestyle"]="dotted"
# change the order in which colors are chosen
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=['#32a852', 'r', 'b', 'y'])
plt.rcParams['font.size']=16
plt.rcParams['axes.facecolor']= 'f0f0f0'
plt.rcParams['font.family'] = 'serif'
plt.rcParams['lines.linewidth']= 5
plt.rcParams['xtick.labelsize'] = 24
plt.rcParams['ytick.labelsize'] = 24
````

````python
## Notice the changes to our simple plots from above
mpl_plot()
````

````python
# NOTICE THE CHANGES OCCUR IN SEABORN TOO, SINCE IT ALSO USES rcParam 
sns_plot()
````

## The `matplotlibrc` file

You can save your configuration (going into `rcParams`) in a file named `matplotlibrc`. 

You can have a global one, stored in `mpl.get_configdir()`

You can also have separate ones per project or folder.

**Advantage**: You can put the configuration under version control and maintain its provenance.

The `matplotlibrc` file would look something like this, similar to a python dictionary

```
axes.axisbelow : True       # Draw axis grid lines and ticks below patches (True); above
                            # patches but below lines ('line'); or above all (False).
                            # Forces grid lines below figures.
font.size : 12              # Font size in pt.
grid.linewidth : 1.2        # In pt.
legend.framealpha : 1       # Legend patch transparency.
legend.scatterpoints : 3    # Number of scatter points in legend.
lines.linewidth : 3         # line width in pt.
```

Notice that the tags are the same as in rcParams, so you can edit the file in a similar way. 


## Create a new _style_

````python
from pathlib import Path
cfgdir = mpl.get_configdir() # find your configuration folder
p = Path(cfgdir)
print(p)
stylelib = (p / 'stylelib')
stylelib.mkdir(exist_ok=True)
path = stylelib / 'mycustomstyle.mplstyle' # create paths
path.write_text(''' # write into the file
axes.facecolor : f0f0f0
font.family : serif
lines.linewidth : 5
xtick.labelsize : 24
ytick.labelsize : 24
''')
````

## Customize subplots/panels

Creating subplots is quite easy in matplotlib using the `subplots` function

You can add a title or other customization to just one of the panel

````python
x = np.linspace(0,100,50)
y = np.random.uniform(low=0, high=10, size=50)
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (10,10/1.618))
ax1.scatter(x,y, c='red', marker='+')
ax2.bar(x,y);
plt.show()
````

````python
fig, ax = plt.subplots(2,2, figsize = (10,12))
ax[0,0].bar(x,y)
ax[1,0].bar(x,y)
ax[0,1].scatter(x,y)
ax[1,1].plot(x)
ax[1,1].set_title('Plot 4')
plt.show()
````

You can also change the padding between the plots

````python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2, figsize=(10,12))
ax1.scatter(x,y,c='red', marker='+')
ax2.bar(x,y)
ax3.scatter(x,y, marker='x')
ax4.barh(x,y)
ax4.set_title('Plot 4')
plt.subplots_adjust(wspace=0.5, hspace=1.0)
plt.show()
````


## Size considerations

The size of a plot on your publication is often central to its aesthetics. This size can be specified in each matplotlib plot. 

An aesthetically pleasing ratio of with to height is the _Golden Ratio_, which is approximately 1.618. We can write a function to ensure that for any particular width we desire, we can set the height to meet the Golden ratio.


````python
def set_size(width):
    """Set aesthetic figure dimensions following golden ratio
    Args:
        width (float): width of the figure in inches (what matplotlib uses)
    Returns:
        fig_dim (tuple): Dimensions of the figure in inches
    """
    golden_ratio = (5 ** 0.5 -1)/2
    fig_height = width * golden_ratio
    return width, fig_height
````

````python
x = np.linspace(0,10,num=1000)
fig, ax = plt.subplots(figsize = set_size(10))
ax.plot(x, np.sin(x))
plt.show()
````
